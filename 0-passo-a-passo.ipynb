# Guia 1: Databricks App + Vibe Coding com Genie Code
## Caso de Uso: Simulador de Score de Financiamento Automotivo

---

## Contexto do Projeto

Um cliente do **setor automotivo** precisa de uma solução para que seus **vendedores simulem o score de financiamento** de veículos com base nos dados do cliente. O app será desenvolvido usando **Databricks Apps com Streamlit**, construído inteiramente via **Genie Code** (vibe coding).

---

## Passo 1 — Criar o Databricks App

1. No menu lateral do Databricks, clique em **Compute** → **Apps**.
2. Clique em **Create App**.
3. Preencha o **App name** com o nome desejado para sua aplicação.
4. Em **App source code path**, selecione o caminho:
   ```
   /Workspace/Users/{seu_usuario}/1-databricks-app-streamlit
   ```
   > ⚠️ Substitua `{seu_usuario}` pelo seu e-mail de usuário no workspace.
5. Clique em **Create App** para confirmar.

---

# Guia 2: Criação e Configuração de um Databricks App API

---

## Passo 1 — Criar o Databricks App

1. No menu lateral do Databricks, clique em **Compute** → **Apps**.
2. Clique em **Create App**.
3. Preencha o **App name** com o nome desejado para sua aplicação.
4. Em **App source code path**, selecione o caminho:
   ```
   /Workspace/Users/{seu_usuario}/app
   ```
   > ⚠️ Substitua `{seu_usuario}` pelo seu e-mail de usuário no workspace.
5. Clique em **Create App** para confirmar.

---

## Passo 2 — Copiar o Service Principal e a URL do Databricks

Após a criação do app, a tela de detalhes exibirá informações importantes:

1. Copie a **URL do Databricks workspace** (formato: `https://<workspace-id>.cloud.databricks.com`).
2. Guarde ambos os valores — serão necessários nos próximos passos.

---

## Passo 3 — Criar um Service Principal e Gerar client_id / client_secret

### O que é um Service Principal?
Um Service Principal é uma identidade de máquina (não-humana) que permite que aplicações se autentiquem no Databricks de forma programática, sem usar credenciais pessoais de usuário.

### Como criar:

1. Acesse **Admin Settings** (ícone de engrenagem no canto superior direito) → **Identity and access** → **Service principals**.
2. Clique em **Add service principal** → **Add new**.
3. Dê um nome descritivo para o Service Principal.
4. Clique em **Add**.

### Como gerar client_id e client_secret (OAuth):

1. Dentro do Service Principal recém-criado, vá para a aba **Secrets**.
2. Clique em **Generate secret**.
3. O sistema exibirá:
   * **Client ID** (UUID do service principal)
   * **Client Secret** (token gerado — exibido **apenas uma vez**)

> 🔴 **IMPORTANTE:** Copie e guarde o `client_id` e o `client_secret` em um local seguro (ex: gerenciador de senhas, vault, ou arquivo protegido). O secret **NÃO** será exibido novamente após fechar a janela.

---

## Passo 4 — Compartilhar o App com o Service Principal

1. Volte para o seu aplicativo: **Compute** → **Apps** → selecione o app criado.
2. Clique no botão **Share** (ou **Permissions**).
3. No campo de busca, digite o nome do Service Principal criado no Passo 3.
4. Selecione-o e atribua a permissão CAN_USE.
5. Clique em **Save** / **Add**.

Isso permite que o app utilize as credenciais do Service Principal para acessar recursos do workspace.

---

## Passo 5 — Testar a API invocando o notebook

1. Abra o notebook **`2-testing-api.ipynb`** no workspace.
2. Certifique-se de que as variáveis de conexão (URL do workspace, client_id, client_secret) estão configuradas corretamente no notebook.
3. Execute o notebook **inteiro** (clique em **Run All** ou use `Ctrl+Shift+Enter`).
4. Verifique os resultados de cada célula para confirmar que a API está respondendo corretamente.
