```python
from databricks.sdk import WorkspaceClient
import requests


CLIENT_ID="{your_id}"
CLIENT_SECRET="{your_secret}"

APP_URL = "{your_app_url}"
WORKSPACE_URL = "{your_workspace_url}"

# IMPORTANTE: Use as credenciais do service principal app-externo-crivo
wc = WorkspaceClient(
    host=f"{WORKSPACE_URL}", 
    client_id=f"{CLIENT_ID}",  # Client ID do app-externo-crivo
    client_secret=f"{CLIENT_SECRET}"  # Client Secret do app-externo-crivo
)

# Gerar Bearer token com as credenciais do service principal
headers = wc.config.authenticate()
token = headers['Authorization']

print(f"Token gerado para o service principal: {token}")

# Fazer chamada à API do Databricks App
response = requests.get(
    f"{APP_URL}/api",
    headers={"Authorization": token}  # Usar o token completo (já inclui "Bearer ")
)

import json
print(f"Status Code: {response.status_code}")
print(f"Response Text: {response.text}")
try:
    print(f"Response JSON: {json.dumps(response.json(), indent=2)}")
except:
    pass
```

```python
# Rota 2: GET /api/modelos — Listar modelos de scoring disponíveis
# Sem filtro (retorna todos)
response = requests.get(
    f"{APP_URL}/api/modelos",
    headers={"Authorization": token}
)
print("=== GET /api/modelos (todos) ===")
print(f"Status: {response.status_code}")
print(json.dumps(response.json(), indent=2, ensure_ascii=False))

# Com filtro por segmento (query parameter)
response_filtrado = requests.get(
    f"{APP_URL}/api/modelos",
    headers={"Authorization": token},
    params={"segmento": "pf"}
)
print("\n=== GET /api/modelos?segmento=pf ===")
print(f"Status: {response_filtrado.status_code}")
print(json.dumps(response_filtrado.json(), indent=2, ensure_ascii=False))
```

```python
# Rota 3: POST /api/selecionar-modelo — Enviar parâmetros no body (JSON)
payload = {
    "cpf_cnpj": "12345678901",
    "valor_veiculo": 85000.00,
    "categoria_veiculo": "suv",
    "segmento": "pf"
}

response = requests.post(
    f"{APP_URL}/api/selecionar-modelo",
    headers={"Authorization": token, "Content-Type": "application/json"},
    json=payload
)
print("=== POST /api/selecionar-modelo ===")
print(f"Payload enviado: {json.dumps(payload, indent=2)}")
print(f"\nStatus: {response.status_code}")
print(f"Resposta:\n{json.dumps(response.json(), indent=2, ensure_ascii=False)}")
```

```python
# Rota 4: GET /api/score/{modelo_id}/{cpf_cnpj} — 2 parâmetros na URL (path params)
modelo_id = "score_credito_pf"
cpf_cnpj = "12345678901"

response = requests.get(
    f"{APP_URL}/api/score/{modelo_id}/{cpf_cnpj}",
    headers={"Authorization": token}
)
print(f"=== GET /api/score/{modelo_id}/{cpf_cnpj} ===")
print(f"Status: {response.status_code}")
print(json.dumps(response.json(), indent=2, ensure_ascii=False))

# Testando com outro modelo
modelo_id_2 = "score_fraude_identidade"
response_2 = requests.get(
    f"{APP_URL}/api/score/{modelo_id_2}/{cpf_cnpj}",
    headers={"Authorization": token}
)
print(f"\n=== GET /api/score/{modelo_id_2}/{cpf_cnpj} ===")
print(f"Status: {response_2.status_code}")
print(json.dumps(response_2.json(), indent=2, ensure_ascii=False))
```
